# Polynomial Regression

***Polynomial Regression*** is a form of regression analysis in which the relationship between the independent variable $x$ and the dependent variable
$y$ is modeled as an $n^{th}$ degree polynomial. It is an extension of linear regression, which models the relationship as a straight line. Polynomial regression can capture the non-linear relationships between the variables, making it a powerful tool for fitting more complex data. In addition, it relies on some key assumptions: Independence, Homoscedasticity, and Normality.

### Equation

The relationship between $x$ and $y$ is modeled by an equation of the form:

$y = \beta_0 + \beta_1x + \beta_2x^2 + \beta_3x^3 + ... + \beta_nx^n + \epsilon$

* $y$ is the dependent variable
* $x$ is the independent variable
* $\beta_0,\beta_1, .. $ are the coefficients of the polynomial
* $\epsilon$ is the error term

## Goodness of Fit
* R-squared ($R^2$): Indicates the proportion of variance in the dependent variables explained by the independent variables. A higher $R^2$ indicates a better fit.
* Adjusted R-squared: Adjusted for the number of predictors in the model, providing a more accurate measure when multiple variables are involved.
* Residuals Analysis: Examining residuals helps check assumptions and identify potential issues.

## Overfitting

One of the risks with Polynomial Regression is overfitting, especially with high-degree polynomials. Overfitting occurs when the model becomes too complex and captures the noise in teh data rather than the underlying pattern. To mitigate overfitting:
* Cross-Validation: Use cross-validation techniques to assess the model's performance on different subsets of data.
* Regularization: Apply regularization techniques (like Ridge or Lasso Regression) to penalize large coefficients and simplify the model.



# Example

Predicting Salary based on the Position

```python
# Import the necessary Libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import Data
data = pd.read_csv('Position_Salaries.csv')
X = data.iloc[:, 1:-1].values
y = data.iloc[:,-1].values
```

```python
# Train the Linear Regression Model on the whole Data

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
# Train the Polynomial Regression model on teh whole Data

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# Visualizing the Linear Regression results

plt.scatter(X, y, color = 'blue')
plt.plot(X, lin_reg.predict(X), color = 'black')
plt.title('Linear Regression - Truth or Bluff')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

```python
# Visualizing the Polynomial Regression results

plt.scatter(X, y, color = 'blue')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'black')
plt.title('Polynomial Regression - Truth or Bluff')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

```python
# Visualising the Polynomial Rgression Results
# This gives higher resolution and smoother curve

X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
```

```python
# Predicting a new result with Linear Regression

lin_reg.predict([[6.5]])
```

```python
# Predicting a new result with Polynomial Regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
