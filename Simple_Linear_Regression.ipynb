# Simple Linear Regression

Simple Linear Regression is a statistical method used to understand the relationship between two continuous variables, *independent* and *dependent*. Independent variable, or predictor, is used to predict the value of the dependent variable, or response. Therefore, the Simple Linear Regression helps to model the linear relationship between the independent variable (X) and the dependent variable (Y).

For Simple Linear Regression to be valid, several assumptions need to be met:

*   Linearity: The relationship between X and Y is linear.
*   Independence: Observations are independent of each other.
*   Homscedasticity: The variance of residual (error) terms is constant across all levels of the independent variable.
*   Normality: The residuals of the model are normally distributed.

The relationship between X and Y is described by the equation of a straight line:

$Y = \beta_0 + \beta_1X + \epsilon$

where,

* $Y$ is the dependent variable
* $X$ is the independent variable
* $\beta_0$ is the intercept of the regression line
* $\beta_1$ is the slope of the regression line. This indicates the change in $Y$ for a one-unit change in $X$. A positive slope indicates a positive relationship, while a engative slope indicates a negative relationship.
* $\epsilon$ is the error term, representing the difference between the observed and the predicted values.

To find the best fitting line, we need to estimate the parameters $\beta_0$ and $\beta_1$. This is typically done by using the Least Squares Method, which minimizes the sum of the squared differences between the observed values and the values predicted by the line.

To evaluate how well the model fits the data, we often look at:
* R-squared ($R^2$): The proprotion of variance in the dependent variable that is predictable from the independent variable. An $R^2$ value closer to 1 indicates a better fit.
* Residuals: The diffferences between the observed and predicted values. Analyzing residuals helps in checkign the assumptions and diagnosing potential issues.


## Example

Predicting employee Salary based on their Years of experience.

**Import necessary libraries**

```python
import numpy
import matplotlib.pyplot as plt
import pandas as pd
```

**Import the data**

```python
data = pd.read_csv('Salary_Data.csv')
X = data.iloc[:, :-1].values
y = data.iloc[:,-1].values
```

**Split the data into the Training set and the Test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

**Train the Simple Linear Regression model**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

**Predict the Test set results**

```python
y_pred = regressor.predict(X_test)
```

**Visualize the Training set results**

```python
plt.scatter(X_train,y_train, color = 'blue')
plt.plot(X_train, regressor.predict(X_train), color='black')
plt.title('Salary vs. Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

***Visualize the Test set results**

```python
plt.scatter(X_test,y_test, color = 'blue')
plt.plot(X_train, regressor.predict(X_train), color='black')
plt.title('Salary vs. Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Make a Single Prediction

```python
print(regressor.predict([[12]]))
```

Get the Intercept and the Coefficients

```python
print(regressor.coef_)
print(regressor.intercept_)
```

The equation of our Simple Linear Regression model is:
